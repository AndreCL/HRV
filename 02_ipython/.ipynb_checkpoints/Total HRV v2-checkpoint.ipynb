# Scripts used to analyse data

```python
#Imports
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Import data finction
def getfile(name):
    tempdict = {}
    with open(name) as f:
            for line in f:
                line = line.replace('\n', '')
                line = line.split(";")
                tempdict[line[0]] = float(line[1])
    #print len(tempdict) #for possible debugging
    return tempdict
```

```python
#Get mean(average) function

#add together
def dsum(*dicts):
    ret = defaultdict(int)
    for d in dicts:
        for e in d:
            for k, v in e.items():
                ret[k] += v
    return dict(ret)

#mean function
def getmean(files):
    amount = len(files)
    finaldict1 = dsum(files) #raw data
    finallist2 = [] #average
    for i in finaldict1:
        finallist2.append(finaldict1[i] / amount)
    return finallist2
    
```

```python
#Get median
def getmedian(files):
    medianlist = []
    for i1 in range(36): #Change for other files
        temp = []
        for i2 in files:
            temp.append(i2.values()[i1])
        medianlist.append(np.median(temp))
    return medianlist
```

## Data analysis

```python
#quick and dirty way
g01 = getfile("2015-11-04andre")
g02 = getfile("2015-11-04jose")
g03 = getfile("2015-11-04mike")
g04 = getfile("2015-11-04trevon")
g05 = getfile("2015-11-06g1")
g06 = getfile("2015-11-06g2")
g07 = getfile("2015-11-06g3")
g08 = getfile("2015-11-06g4")
g09 = getfile("2015-11-06g5")
g10 = getfile("2015-11-06g6")
g11 = getfile("2015-11-11g7")
g12 = getfile("2015-11-11g8")
g13 = getfile("2015-11-11g9")
g14 = getfile("2015-11-11thomas")
```

```python
total = [g01,g02,g03,g04,g05,g06,g07,g08,g09,g10,g11,g12,g13,g14]
```

```python
%matplotlib inline
#GENERAL

plt.figure(1)

plt.subplot(211)

#all dots
for i in total:
    temp = []
    for i2 in i:
        temp.append(i[i2])
    plt.plot(temp, 'r.')
    
plt.axvspan(9, 15.8, facecolor='y', alpha=1)
plt.axvspan(17.6, 21.85, facecolor='y', alpha=1)
plt.axvspan(23.25, 28.7, facecolor='y', alpha=1)
    
plt.xlim((0,36))
plt.title('Raw')
plt.subplot(212)

#average
line_one, = plt.plot(getmean(total), 'g', label = 'mean')
line_two, = plt.plot(getmedian(total), 'b', label = 'median')

plt.ylim((0,7))
plt.xlim((0,36))

plt.title('Results')
plt.legend(bbox_to_anchor=(1, 1), handles=[line_one, line_two],prop={'size':6})
plt.ylabel('LF/HF')

plt.axvspan(9, 15.8, facecolor='y', alpha=1)
plt.axvspan(17.6, 21.85, facecolor='y', alpha=1)
plt.axvspan(23.25, 28.7, facecolor='y', alpha=1)


plt.tight_layout()

plt.savefig('graph1.png', dpi=1000, transparent=True)
```
